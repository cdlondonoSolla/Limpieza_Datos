#### La curtosis es una medida estadística que describe la forma de la distribución de los datos, particularmente en cuanto a las colas y la concentración de los valores en el centro. 
#### Nos ayuda a identificar si los datos tienen valores extremos con mayor o menor frecuencia que una distribución normal.

### Tipos de Curtosis
#### 1️⃣ Curtosis Mesocúrtica (Normal)

####  La curtosis de una distribución normal es 0 (usando la definición de Fisher).
####  Tiene una distribución de datos equilibrada con colas similares a la normal.

### 2️⃣ Curtosis Leptocúrtica (>0)

#### Presenta colas pesadas, lo que significa que hay más valores extremos (outliers).
#### Ejemplo: Distribuciones de tipo Laplace o t-Student con pocos grados de libertad.

### 3️⃣ Curtosis Platicúrtica (<0)

#### Tiene colas ligeras, lo que indica menos valores extremos.
#### Ejemplo: Distribución uniforme, donde los valores están más dispersos.Tipos de Curtosis

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurar estilo de gráficos
sns.set(style="darkgrid")
```

```python
# Generar datos
# 1. Distribución Normal (Mesocúrtica, Curtosis ≈ 0)
data_normal = np.random.normal(0, 1, 1000)
```

```python
# 2. Distribución Laplace (Leptocúrtica, Curtosis > 0, colas pesadas)
data_leptokurtic = np.random.laplace(0, 1, 1000)
```

```python
# 3. Distribución Uniforme (Platicúrtica, Curtosis < 0, colas ligeras)
data_platykurtic = np.random.uniform(-2, 2, 1000)
```

```python
# Calcular la curtosis de cada conjunto de datos
curtosis_normal = stats.kurtosis(data_normal, fisher=True)
curtosis_lepto = stats.kurtosis(data_leptokurtic, fisher=True)
curtosis_platy = stats.kurtosis(data_platykurtic, fisher=True)
```

```python

# Crear la figura y los ejes para los gráficos
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Listado de datasets con etiquetas
datasets = [
    (data_normal, curtosis_normal, "Normal (Mesocúrtica, ≈0)"),
    (data_leptokurtic, curtosis_lepto, "Leptocúrtica (>0)"),
    (data_platykurtic, curtosis_platy, "Platicúrtica (<0)")
]

# Generar gráficos de distribución
for ax, (data, kurt, title) in zip(axes, datasets):
    sns.histplot(data, kde=True, bins=30, ax=ax, color='royalblue')
    ax.set_title(f"{title}\nCurtosis = {kurt:.2f}")
    ax.set_xlabel("Valor")
    ax.set_ylabel("Frecuencia")

# Ajustar el diseño del gráfico
plt.suptitle("Comparación de Distribuciones según la Curtosis", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Crear un boxplot para comparar las distribuciones
plt.figure(figsize=(10, 6))
sns.boxplot(data=[data_normal, data_leptokurtic, data_platykurtic], 
            palette=["blue", "red", "green"])
plt.xticks([0, 1, 2], ["Normal", "Leptocúrtica", "Platicúrtica"])
plt.title("Comparación de Distribuciones con Boxplot")
plt.ylabel("Valor")
plt.show()
```
