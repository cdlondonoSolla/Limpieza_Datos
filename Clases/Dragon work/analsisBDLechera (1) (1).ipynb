## Adicionar en el data set columna con el mes y año

```python
# Crear la nueva columna con mes y año
df['Mes_Año'] = df['Fe.contabilización'].dt.strftime('%m %Y')

# Mostrar el DataFrame resultante
print(df.head())
```

```python
## preguntar por que adiciona fecha y mes al final
```

## Se agrupa por numero de Usuario encargado de realizar ajustes por el tipo de movimiento aprobado para tal fin

```python
# Filtrar el DataFrame para el usuario User078 y las clases de movimiento 701 y 702
filtered_df = df[(df['Nombre del usuario'] == 'User078') & 
                  (df['Clase de movimiento'].isin([701, 702]))]

# Agrupar por las columnas y sumar la cantidad e importe
grouped_df = filtered_df.groupby(['Material', 'Texto breve de material', 'Centro', 'Almacén', 'Mes_Año']).agg({
    'Cantidad': 'sum',
    'Importe ML': 'sum'
}).reset_index()

# Mostrar el resultado
print(grouped_df)
```

## Agrupación programación de inventario

```python
Cronograma = pd.read_excel("Cronograma de inventarios.xlsx")
```

```python
Cronograma.info()
```

```python
Cronograma['Fecha'] = pd.to_datetime(Cronograma['Fecha'])

# Crear la nueva columna 'Mes_Año' combinando mes y año
Cronograma['Mes_Año'] = Cronograma['Fecha'].dt.strftime('%m %Y')  

# Mostrar el DataFrame actualizado
print(Cronograma)
```

## Aqui debe hacer un condicional que si en ajustes de inventarios esta el mes me lo compare con el cronograma

```python
# Crear una columna de coincidencia en Cronograma
Cronograma ['Coincidencia'] = Cronograma.apply(
    lambda row: 1 if ((row['Material'], row['Mes_Año']) in 
                      zip(grouped_df ['Material'], grouped_df ['Mes_Año'])) else 0, axis=1
)

# Mostrar el DataFrame resultante
print(Cronograma)
```

```python
# Filtrar el DataFrame para obtener solo los datos a partir de enero de 2024
Cronograma['Mes_Año'] = pd.to_datetime(Cronograma['Mes_Año'])
Cronograma_filtered = Cronograma[Cronograma['Mes_Año'] >= '2024-01-01']

# Agrupar los datos por Mes y Material, y contar los cumplimientos
cumplimientos = Cronograma_filtered.groupby([Cronograma_filtered['Mes_Año'].dt.to_period('M'), 'Material'])['Coincidencia'].sum().reset_index()

# Convertir la columna de periodo a fecha para facilitar la graficación
cumplimientos['Mes_Año'] = cumplimientos['Mes_Año'].dt.to_timestamp()

# Graficar
plt.figure(figsize=(12, 6))
for material in cumplimientos['Material'].unique():
    data = cumplimientos[cumplimientos['Material'] == material]
    plt.plot(data['Mes_Año'], data['Coincidencia'], marker='o', label=material)

plt.title('Cumplimientos por Mes a partir de Enero 2024')
plt.xlabel('Mes')
plt.ylabel('Cumplimientos')
plt.xticks(rotation=45)
plt.legend(title='Material')
plt.grid()
plt.tight_layout()
plt.show()
```

## Se adiciona a la data final el cronograma.

```python
Final=grouped_df.merge(Cronograma, how="inner", on="Material")
```
