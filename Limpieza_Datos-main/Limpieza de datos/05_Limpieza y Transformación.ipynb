## Transformación de datos

### 2da Opción: Reemplazar NaNs por la media

### Parsing ("Parseo") de la columna Salary

```python
df_sin_nulos['salary']
```

```python
## Este código toma una columna `'salary'` que contiene valores de salario en formato de cadena, 
## elimina cualquier caracter no numérico, extrae los dígitos y convierte la columna resultante en valores numéricos de tipo float. 
## El resultado se almacena en una nueva columna llamada `'salary_numeric'`. 
## Esto es útil cuando la columna `'salary'` inicial contiene caracteres no numéricos y se desea trabajar con los salarios como valores numéricos.

df_sin_nulos['salary_numeric'] = df_sin_nulos['salary'].replace('[^\d.]', '', regex=True).str.extract('(\d+)').astype(float)
```

### Método de reemplazo de valores nulos con la media de la columna

```python
# Rellenar valores nulos con un valor específico (por ejemplo, el promedio)
df_con_relleno = df_sin_nulos.fillna(df_sin_nulos['salary_numeric'].mean())
```

```python
# 1. Definir rangos de salario
bins = [0, 30000, 60000, 90000, 120000, 150000]  # Límites de los rangos salariales
labels = ['0-30K', '30K-60K', '60K-90K', '90K-120K', '120K-150K']  # Etiquetas para cada rango

# 2. Asignar cada salario a un rango
df_sin_nulos['salary_range'] = pd.cut(df_sin_nulos['salary_numeric'], bins=bins, labels=labels)

# 3. Contar cuántos valores hay en cada categoría
salary_counts = df_sin_nulos['salary_range'].value_counts()

# 4. Crear la gráfica circular
plt.figure(figsize=(8, 8))  # Ajustar el tamaño del gráfico

# Definir colores personalizados
colors = ['blue', 'green', 'red', 'purple', 'orange']

# Crear el gráfico de pastel con etiquetas y porcentajes
wedges, texts, autotexts = plt.pie(
    salary_counts, 
    labels=salary_counts.index,  
    autopct='%1.1f%%',  # Mostrar porcentaje con un decimal
    colors=colors,  # Colores personalizados
    startangle=90,  # Inicia la gráfica desde arriba (90°)
    pctdistance=0.8,  # Ajusta la posición de los porcentajes
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}  # Borde negro en los sectores
)

# 5. Agregar la leyenda
plt.legend(wedges, labels, title="Rangos Salariales", loc="best")

# 6. Agregar título
plt.title('Distribución de Salarios')

# 7. Mostrar el gráfico
plt.show()
```

### Agrupar y agregar datos

```python
# Supongamos que tienes un DataFrame llamado df con columnas 'title' y 'salary'
# Agrupar por título y calcular el promedio y la mediana del salario para cada título
df_grouped = df_con_relleno.groupby('title')['salary_numeric'].agg(['mean', 'median', 'count']).reset_index()

# Renombrar las columnas para mayor claridad
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'job_count']

# Mostrar el DataFrame resultante
df_grouped
```

```python
df_grouped['salary_mean'] = df_grouped['salary_mean'].round(0).astype(int)
df_grouped['salary_median'] = df_grouped['salary_median'].round(0).astype(int)
```
