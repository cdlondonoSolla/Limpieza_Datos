```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python


# Crear datos simulados usando un diccionario
data = {
    'salario': np.random.randint(20000, 100000, 50),   # Salario en dólares
    'experiencia': np.random.randint(1, 20, 50),       # Años de experiencia
    'edad': np.random.randint(22, 60, 50),             # Edad en años
    'nivel_educativo': np.random.randint(1, 5, 50)     # Nivel educativo (1=Secundaria, 2=Universidad, etc.)
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Calcular la matriz de correlación
corr = df.corr()

# Crear el mapa de calor
plt.figure(figsize=(8, 6))  # Tamaño del gráfico
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Agregar título y mostrar el gráfico
plt.title('Mapa de Calor de Correlación entre Variables')
plt.show()
```
